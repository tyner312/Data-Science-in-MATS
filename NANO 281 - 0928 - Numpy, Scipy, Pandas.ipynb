```python
# Part 2
import numpy as np
x = np.array([[1, 2, 3],
              [4, 7, 6],
              [9, 4, 2]])
y = np.array([1.5, 0.5, 3])
print(x)
print(y)
```

```python
print(x+x)
print(x*x)
print(y*y)
```

```python
# Inner product
print(np.dot(x, x))
```

```python
print(np.dot(x, y))
```

```python
x @ y
```

```python
# Linear algebra, inverse of x
np.linalg.inv(x)
```

```python
# Eigen values and vectors
print(np.linalg.eig(x))
```

```python
# Random 1000 numbers, from 0 to 1
r = np.random.rand(1000, 1)
print(r)
```

```python
# Magic function time
%time sum([i ** 2 for i in r]) ** 0.5
```

```python
# This method should be much faster
%time np.sqrt(np.sum(r**2))
```
